```python
import numpy as np
import tensorflow as tf
```

```python
x = np.random.randn(2,4,5)
x[1, 1:] = 0
seq_lengths = [4,1]
print('x:\n',x)
```

```python
cell = tf.contrib.rnn.BasicLSTMCell(num_units=5, state_is_tuple=True, reuse=True)
outputs, last_states = tf.nn.dynamic_rnn(cell, x, seq_lengths, dtype=tf.float64)
```
